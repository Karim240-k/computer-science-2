```python
# Step 1: Define login credentials and initial balance
USERNAME = "user123"
PASSWORD = "pass123"
balance = 1000.0  # Initial amount of money in the account

# Step 2: Set up login attempts and login status
login_attempts = 0         # Counter for failed attempts
max_attempts = 3           # Maximum number of allowed attempts
logged_in = False          # Boolean flag to track login status

# Step 3: User Login - the system allows 3 attempts
while login_attempts < max_attempts:
    username_input = input("Enter username: ")         # Ask for username
    password_input = input("Enter password: ")         # Ask for password

    if username_input == USERNAME and password_input == PASSWORD:
        print("Login successful!\n")                   # If credentials match, login is successful
        logged_in = True                               # Update login status
        break                                           # Exit login loop
    else:
        login_attempts += 1                            # Increase attempt count
        print(f"Invalid credentials. Attempts left: {max_attempts - login_attempts}\n")

# Step 4: End program if login fails 3 times
if not logged_in:
    print("Too many failed attempts. Exiting system.")  # Notify user
    exit()                                              # Exit the program

# Step 5: Show Main Menu repeatedly until user chooses to exit
while True:
    print("\n--- Main Menu ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Choose an option (1-4): ")  # Get user's menu choice

    # Step 6: Option 1 - Check Balance
    if choice == "1":
        print(f"Your current balance is: ${balance:.2f}")

    # Step 7: Option 2 - Deposit Money
    elif choice == "2":
        try:
            amount = float(input("Enter amount to deposit: "))  # Get deposit amount
            if amount > 0:
                balance += amount
                print(f"${amount:.2f} deposited successfully.")  # Show confirmation
            else:
                print("Invalid amount. Please enter a positive number.")  # Negative or zero input
        except ValueError:
            print("Invalid input. Please enter a number.")  # If input is not a number

    # Step 8: Option 3 - Withdraw Money
    elif choice == "3":
        try:
            amount = float(input("Enter amount to withdraw: "))  # Get withdrawal amount
            if amount > balance:
                print("Insufficient funds.")  # Can't withdraw more than current balance
            elif amount <= 0:
                print("Invalid amount. Please enter a positive number.")  # Negative or zero input
            else:
                balance -= amount
                print(f"${amount:.2f} withdrawn successfully.")  # Show confirmation
        except ValueError:
            print("Invalid input. Please enter a number.")  # If input is not a number

    # Step 9: Option 4 - Exit the system
    elif choice == "4":
        print("Thank you for using the banking system. Goodbye!")  # Goodbye message
        break  # Exit the main loop

    # Step 10: Handle Invalid Menu Choices
    else:
        print("Invalid option. Please choose between 1 and 4.")
```
