```python
# Define the correct login credentials and set how many attempts are allowed
username = "user1"
password = "pass123"
tries = 3

# Keep asking for login details as long as the user has tries left
while tries > 0:
    user_input = input("Enter username: ")  # Prompt the user to type in their username
    pass_input = input("Enter password: ")  # Prompt the user to type in their password

    if user_input == username and pass_input == password:
        # If the login details are correct
        print("Login successful!")
        break  # Exit the loop immediately after a successful login
    else:
        # If the login details are incorrect
        tries = tries - 1  # Reduce the remaining attempts
        print("Wrong username or password!")
        print("Tries left:", tries)

# Once the loop ends, check if the user used up all their attempts
if tries == 0:
    print("Too many wrong tries Program will stop!")  # Inform the user they’ve been locked out

# If the login was successful, let the user access the banking system
if tries != 0:
    balance = 1000  # Set the starting balance

    # Keep showing the menu until the user chooses to exit
    while True:
        print("")
        print("----- MAIN MENU -----")
        print("1. Check balance")
        print("2. Deposit money")
        print("3. Withdraw money")
        print("4. Exit")

        choice = input("Please choose 1, 2, 3, or 4: ")  # Ask the user for a menu option

        if choice == "1":
            # Show the current account balance
            print("Your balance is:", balance)

        elif choice == "2":
            # Let the user add money to their account
            deposit = input("How much do you want to deposit?: ")
            deposit = float(deposit)
            if deposit > 0:
                balance = balance + deposit  # Update the balance
                print("Money added! Your new balance is:", balance)
            else:
                print("Please enter a number more than 0!")

        elif choice == "3":
            # Let the user take out money
            withdraw = input("How much do you want to withdraw?: ")
            withdraw = float(withdraw)
            if withdraw > 0 and withdraw <= balance:
                balance = balance - withdraw  # Update the balance
                print("Money taken out! Your new balance is:", balance)
            else:
                print("Not enough money or bad number!")

        elif choice == "4":
            # End the session if the user selects this option
            print("Goodbye! Thank you for using the bank!")
            break  # Exit the loop and stop the program

        else:
            # Handle invalid menu selections
            print("That is not a choice! Please pick 1, 2, 3, or 4.")

# Notes:
# - The menu keeps appearing because of the infinite loop (while True).
# - The loop only ends when the user chooses option 4 (Exit), which triggers the break statement.
```

```python
# Project: Movie Library System

# Step 1: Create the Movie Library
# Create an empty list to store all movie dictionaries
movie_library = []

# Step 2: Add Movies to the Library as dictionaries

# Define the first movie as a dictionary with key information
movie1 = {
    "Title": "Inception",
    "Year": 2010,
    "Genres": ["Action", "Sci-Fi"],
    "Rating": (4.5, 1500),  # (average rating, number of reviews)
    "Reviews": {
        "Alice": "Mind-blowing!",
        "Bob": "Amazing visuals and story."
    }
}

# Define the second movie
movie2 = {
    "Title": "Titanic",
    "Year": 1997,
    "Genres": ["Romance", "Drama"],
    "Rating": (4.7, 2000),
    "Reviews": {
        "Carol": "A beautiful love story.",
        "Dave": "Very emotional and well made."
    }
}

# Define the third movie
movie3 = {
    "Title": "The Matrix",
    "Year": 1999,
    "Genres": ["Action", "Sci-Fi"],
    "Rating": (4.8, 1800),
    "Reviews": {
        "Eve": "Changed sci-fi movies forever!",
        "Frank": "Incredible action scenes."
    }
}

# Add all movies to the movie_library list
movie_library.append(movie1)
movie_library.append(movie2)
movie_library.append(movie3)

# Step 3: View All Movies in the Library
# Define a function to display all movies
def view_all_movies():
    print("\nAll Movies in the Library:")
    for movie in movie_library:
        print("Title:", movie["Title"])
        print("Year:", movie["Year"])
        print("Genres:", movie["Genres"])
        print("Rating:", movie["Rating"][0])
        print("Number of Reviews:", movie["Rating"][1])
        print("-----------------------------")

# Step 4: Search Movies by Genre
# Define a function to search and display movies matching a genre
def search_by_genre():
    genre = input("\nEnter a genre to search: ")
    print("\nMovies in genre:", genre)
    found = False  # Track if any matching movie is found
    for movie in movie_library:
        if genre in movie["Genres"]:
            print("Title:", movie["Title"])
            found = True
    if not found:
        print("No movies found in this genre.")

# Step 5: View Details of a Specific Movie
# Define a function to display full details of a movie by its title
def view_movie_details():
    title = input("\nEnter the movie title: ")
    found = False
    for movie in movie_library:
        if movie["Title"].lower() == title.lower():  # Case-insensitive match
            print("\nMovie Details:")
            print("Title:", movie["Title"])
            print("Year:", movie["Year"])
            print("Genres:", movie["Genres"])
            print("Rating:", movie["Rating"][0])
            print("Number of Reviews:", movie["Rating"][1])
            print("Reviews:")
            for reviewer, review_text in movie["Reviews"].items():
                print(reviewer + ":", review_text)
            found = True
            break  # Stop once the movie is found
    if not found:
        print("Movie not found.")

# Step 6: Add a Review to a Movie
# Define a function to add a new review and update the average rating
def add_review():
    title = input("\nEnter the movie title to add a review: ")
    for movie in movie_library:
        if movie["Title"].lower() == title.lower():
            reviewer_name = input("Enter your name: ")
            review_text = input("Enter your review: ")
            movie["Reviews"][reviewer_name] = review_text  # Add to review dictionary

            # Assume all new reviews give a rating of 5 stars
            old_score, old_reviews = movie["Rating"]
            new_total_score = (old_score * old_reviews) + 5
            new_total_reviews = old_reviews + 1
            new_score = new_total_score / new_total_reviews
            movie["Rating"] = (new_score, new_total_reviews)  # Update rating
            print("Review added successfully!")
            return  # End function after adding
    print("Movie not found.")

# Step 7: Calculate and Display the Average Rating
# Define a function to calculate the overall average rating
def average_rating():
    total_score = 0
    total_reviews = 0
    for movie in movie_library:
        score, reviews = movie["Rating"]
        total_score += score * reviews  # Weighted total
        total_reviews += reviews
    if total_reviews > 0:
        average = total_score / total_reviews
        print("\nAverage Rating of All Movies:", round(average, 2))
    else:
        print("\nNo reviews yet.")

# Step 8: Test Your Program
# Create a simple loop-based menu to interact with the program
while True:
    print("\nMovie Library System")
    print("1. View all movies")
    print("2. Search movies by genre")
    print("3. View movie details")
    print("4. Add a review to a movie")
    print("5. Show average rating")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        view_all_movies()
    elif choice == "2":
        search_by_genre()
    elif choice == "3":
        view_movie_details()
    elif choice == "4":
        add_review()
    elif choice == "5":
        average_rating()
    elif choice == "6":
        print("Goodbye!")
        break  # Exit the program loop
    else:
        print("Invalid choice. Please try again.")
```

```python
# Start with an empty list to keep track of all the transactions
transactions = []

# This function lets the user add a new transaction (either income or expense)
def add_transaction():
    print("\nAdd a New Transaction")

    # Ask the user to fill in the details
    category = input("Enter category (e.g., Food, Rent, Salary): ")
    type_ = input("Enter type (Income/Expense): ").strip().capitalize()
    amount = float(input("Enter amount: "))
    note = input("Enter a note (optional): ")

    # Make sure the user enters either "Income" or "Expense"
    if type_ not in ["Income", "Expense"]:
        print("Invalid type. Please enter 'Income' or 'Expense'.")
        return

    # Store the transaction details in a dictionary
    transaction = {
        "Category": category,
        "Type": type_,
        "Amount": amount,
        "Note": note
    }

    # Add the transaction to the list
    transactions.append(transaction)
    print("Transaction added successfully!")

# This function shows all the transactions you've added so far
def view_transactions():
    print("\nAll Transactions:")

    # If there are no transactions, let the user know
    if not transactions:
        print("No transactions recorded.")
        return

    # Loop through and print each one nicely
    for i, t in enumerate(transactions, start=1):
        print(f"{i}. {t['Type']} - {t['Category']} - ${t['Amount']:.2f} | Note: {t['Note']}")

# This one calculates your total income, expenses, and balance
def calculate_balance():
    # Add up all income and expenses
    income = sum(t["Amount"] for t in transactions if t["Type"] == "Income")
    expense = sum(t["Amount"] for t in transactions if t["Type"] == "Expense")
    balance = income - expense

    # Show a summary so the user knows where they stand
    print("\n💰 Budget Summary")
    print(f"Total Income: ${income:.2f}")
    print(f"Total Expenses: ${expense:.2f}")
    print(f"Remaining Balance: ${balance:.2f}")

# This function helps the user find transactions by category (like just "Food" or "Rent")
def filter_by_category():
    category = input("\nEnter category to filter: ")

    # Filter the list to only show matching transactions
    filtered = [t for t in transactions if t["Category"].lower() == category.lower()]

    # If nothing matches, let the user know
    if not filtered:
        print("No transactions in this category.")
        return

    # Otherwise, display the filtered results
    print(f"\nTransactions in category '{category}':")
    for t in filtered:
        print(f"{t['Type']} - ${t['Amount']:.2f} | Note: {t['Note']}")

# This is the main loop that keeps the program running and asks the user what they want to do
while True:
    print("\n🔹 Personal Budget Tracker 🔹")
    print("1. Add a transaction")
    print("2. View all transactions")
    print("3. Show balance")
    print("4. Filter by category")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    # Match the user’s choice with the correct function
    if choice == "1":
        add_transaction()
    elif choice == "2":
        view_transactions()
    elif choice == "3":
        calculate_balance()
    elif choice == "4":
        filter_by_category()
    elif choice == "5":
        # End the program
        print("Goodbye!")
        break
    else:
        # In case they enter something invalid
        print("Invalid choice. Try again.")
```

```python
# First, let's set up the menu for the restaurant
# It's a dictionary where the key is the food item and the value is its price
menu = {
    "Burger": 5.99,
    "Pizza": 8.99,
    "Salad": 4.99,
    "Fries": 2.99,
    "Soda": 1.99
}

# This list will store all the items that the user orders
order = []

# Function to display the menu so users know what they can choose from
def view_menu():
    print("\n🍽️  Restaurant Menu")
    for item, price in menu.items():
        print(f"{item}: ${price:.2f}")  # shows item and price nicely formatted

# Function to let the user add food items to their order
def add_to_order():
    item = input("\nEnter the item you want to order: ").title()

    # Check if the item exists in the menu
    if item in menu:
        quantity = int(input("Enter quantity: "))
        order.append((item, quantity))  # Store the item and quantity as a tuple
        print(f"✅ {quantity} x {item} added to your order.")
    else:
        print("❌ Item not found on the menu. Please check the spelling or view the menu again.")

# Function to view everything the user has ordered so far
def view_order():
    print("\n🧾 Your Order:")
    if not order:
        print("You haven't ordered anything yet.")
        return

    total = 0
    for item, quantity in order:
        price = menu[item] * quantity  # total price for this item
        total += price
        print(f"{quantity} x {item} = ${price:.2f}")

    print(f"\nTotal so far: ${total:.2f}")

# This function will 'complete' the order and show the final bill
def checkout():
    print("\n🧾 Final Bill:")
    if not order:
        print("You didn't order anything.")
        return

    total = 0
    for item, quantity in order:
        item_total = menu[item] * quantity
        total += item_total
        print(f"{quantity} x {item} = ${item_total:.2f}")

    print(f"\n🟢 Total to Pay: ${total:.2f}")
    print("Thank you for your order! Have a great meal 😊")

# Main menu loop that lets the user interact with the system
while True:
    print("\n🔹 Welcome to the Restaurant Ordering System 🔹")
    print("1. View Menu")
    print("2. Add to Order")
    print("3. View Order")
    print("4. Checkout")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    # Based on user's choice, call the right function
    if choice == "1":
        view_menu()
    elif choice == "2":
        add_to_order()
    elif choice == "3":
        view_order()
    elif choice == "4":
        checkout()
    elif choice == "5":
        print("Goodbye! 🍽️")
        break  # end the loop and exit
    else:
        print("Invalid choice. Please enter a number from 1 to 5.")
```

```python
# Let's start by importing pandas and giving it a nickname 'pd' (pretty standard)
import pandas as pd

# Here's a small dataset we’ll work with – just some names, ages, cities, and sales numbers
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
    "Sales": [250, 400, 300, 500, 450]
}

# Now we’ll turn that dictionary into a DataFrame – this is basically like a table or spreadsheet
df = pd.DataFrame(data)

# Let’s see what the whole table looks like
print("📋 Full DataFrame:")
print(df)

# If you only want to preview the first few rows, this is how you do it (super helpful for big datasets)
print("\n👀 First 3 rows:")
print(df.head(3))

# This gives you a quick overview of the structure – what columns there are, types of data, etc.
print("\n🧾 DataFrame info:")
print(df.info())

# Want to get some stats like averages, min, max, etc.? This line gives you that for all the number columns
print("\n📊 Summary stats:")
print(df.describe())

# Let’s say we want to know the average age of the people in our table
average_age = df["Age"].mean()
print(f"\n🧠 Average Age: {average_age:.1f}")

# Now let’s filter the data – show only the people who made more than 300 in sales
print("\n🔍 People who made more than 300 in sales:")
high_sales = df[df["Sales"] > 300]
print(high_sales)

# We can also sort the table so we see who made the most sales at the top
print("\n⬇️ Sorting by Sales (highest to lowest):")
sorted_df = df.sort_values(by="Sales", ascending=False)
print(sorted_df)

# Let’s add a new column called “Bonus” – it’s 10% of their sales
# This is just to show how you can do simple calculations with DataFrame columns
df["Bonus"] = df["Sales"] * 0.10
print("\n💰 Added a 'Bonus' column (10% of Sales):")
print(df)

# Finally, let’s save the updated table as a CSV file so we can use it later
# Note: This only works if you're running this on your own computer, not online
df.to_csv("sales_data.csv", index=False)
print("\n📁 Data saved to 'sales_data.csv'")
```
